```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Data Files/Mall_Customers.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,3:]
x.head()
```

```python
plt.title("UC Data")
plt.xlabel("AI")
plt.ylabel("SS")
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
sse = []
for k in range(1, 20):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title("EM")
plt.xlabel("K")
plt.ylabel("SSE")
plt.xticks(range(1, 20))
plt.plot(range(1, 20), sse, marker=".", color="b")
```

```python
from sklearn.metrics import silhouette_score
silh = []
for k in range(2,16):
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(x)
    score = silhouette_score(x, labels)
    silh.append(score)

silh
```

```python
plt.title("S")
plt.xlabel("K")
plt.ylabel("SS")
plt.grid()
plt.xticks(range(2, 16))
plt.bar(range(2,16),silh, color="b")
```

```python
km = KMeans(n_clusters=5)
labels = km.fit_predict(x)
```

```python
plt.title("UC")
plt.xlabel("AN")
plt.ylabel("SS")
plt.scatter(x["Annual Income (k$)"], x["Spending Score (1-100)"])
```

```python

plt.title("UC")
plt.xlabel("AN")
plt.ylabel("SS")
plt.scatter(x["Annual Income (k$)"], x["Spending Score (1-100)"], c=labels)
```

```python
alg = AgglomerativeClustering(n_clusters=5)
alabels = alg.fit_predict(x)
```

```python
plt.title("UC")
plt.xlabel("AN")
plt.ylabel("SS")
plt.scatter(x["Annual Income (k$)"], x["Spending Score (1-100)"], c=alabels)
```
