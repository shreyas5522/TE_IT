```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Data Files/temperatures.csv')
```

```python
df.head()
```

```python
x = df['YEAR']
y = df['ANNUAL']
```

```python
plt.title("Temperature Plot of INDIA")
plt.xlabel("Year")
plt.ylabel("Annual Average Temperature")
plt.scatter(x, y)
```

```python
x.shape
```

```python
x = x.values
```

```python
x = x.reshape(117, 1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2030]])
```

```python
predicted = regressor.predict(x)
```

```python
# MAE
np.mean( abs(y - predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, predicted)
```

```python
# MsE
np.mean((y - predicted) ** 2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, predicted)
```

```python
from sklearn.metrics import r2_score
r2_score(y, predicted)
```

```python
regressor.score(x, y)
```

```python
plt.title("Temperature Plot of INDIA")
plt.xlabel("Year")
plt.ylabel("Anuual Average Temperature")
plt.scatter(x, y, label = 'actual', color = 'r')
plt.plot(x, predicted, label = 'predicted', color='b')
plt.legend()
```
